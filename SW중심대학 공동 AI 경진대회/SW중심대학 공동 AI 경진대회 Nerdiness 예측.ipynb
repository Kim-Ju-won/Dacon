## 2.1.1 XGboost

- Xgboost는 앙상블 기법중 Boosting 기법 중 하나로 Gradient Boosting 기법에 속합니다. Gradient Boosting 방식은 에러를 지속적으로 학습하기 때문에 과적합이 될 가능성이 높습니다. 하지만, XGBoost는 Regularization Term(Ridge, Rasso)이 추가되어 있어 과적합을 줄여줄 수 있습니다. Regularization term을 이용해 복잡한 모델에 패널티를 주게 됩니다.!


> Boosting이란? 앙상블 기법중 하나로오분류된 데이터에 초점을 맞추어 더 많은 가중치를 주는 방식입니다.
<br/>초기에는 모든 데이터가 동일한 가중치를 가지지만, 각 round가 종료된 후 가중치와 중요도를 계산하며, 복원 추출 시에 가중치 분포를 더 많이 고려합니다. <br/>Boosting 기법에는 여러가지 기법이 있지만 그 중에서 Gradient Boosting 기법 중 하나인 **XGboost** 기법을 이용해서 모델링을 진행해주었습니다. 

>Gradient Boosting이란?<br/>Round의 합성 분류기의 데이터 별 오류를 예측하는 약한 분류기를 학습하는 방식으로 쉽게 말해 줄일 수 있는 오차를 학습하여 오차를 줄여나가는 방식입니다. 


```python
xgb_model = xgboost.XGBClassifier(learning_rate=0.2,
                                 n_estimators=1000,
                                 max_depth=12,
                                 min_child_weight=1,
                                 gamma=0,
                                 colsample_bytree = 0.7,
                                 subsample=0.75,
                                 objective= 'reg:logistic',
                                 nthread=-1,
                                 reg_alpha = 1e-5,
                                 scale_pos_weight=1,
                                 seed=2011)
xgb_model.fit(train_x,train_y)
predict = xgb_model.predict(train_x)
print(xgb_model.predict_proba)
```

### Xgboost는 변수별 중요도 파악
- Feature importance를 파악할 수 있는 함수를 제공합니다. 해당기능을 통해 어떤 변수가 영향력이 큰지 파악해보도록 합시다. 
  - Gain : 변수별로 손실함수를 줄여준 정도
  - Weight : 데이터를 분리하는데 활용한 횟수
  - Cover : Feature과 관련된 샘플의 갯수
- 단점 : 관점이 여러개라 해석이 불가능함. (consistency가 낮음)

```python
xgboost.plot_importance(xgb_model, importance_type = 'gain', title='gain', xlabel='', grid=False)
xgboost.plot_importance(xgb_model, importance_type = 'cover', title='cover', xlabel='', grid=False)
xgboost.plot_importance(xgb_model, importance_type = 'weight', title='weight', xlabel='', grid=False)
xgboost.plot_importance(xgb_model, importance_type = 'total_gain', title='total_gain', xlabel='', grid=False)
xgboost.plot_importance(xgb_model, importance_type = 'total_cover', title='total_cover', xlabel='', grid=False)
plt.tight_layout()
plt.show()
```

### Shap Value

- Shap value 는 consistency를 보장함
- 특정 관측치 별로 중요도를 볼 수 있음
- 다양한 변수들의 조합을 통해 시각화를 할 수 있어 파악하기 쉬움

```python
explainer = shap.TreeExplainer(xgb_model) # 트리 모델 Shap Value 계산 객체 지정
shap_values = explainer.shap_values(train_x)
```

```python
shap.initjs() # 자바스크립트 초기화 (그래프 초기화)
shap.force_plot(explainer.expected_value, shap_values[1,:], train_x.iloc[1,:]) 
# 첫 번째 검증 데이터 인스턴스에 대해 Shap Value를 적용하여 시각화
# 빨간색이 영향도가 높으며, 파란색이 영향도가 낮음
```

```python
shap.summary_plot(shap_values, train_x)
```

```python
shap.summary_plot(shap_values, train_x, plot_type = "bar") # 각 변수에 대한 Shap Values의 절대값으로 중요도 파악
```

## 하이퍼 파라미터 튜팅 

- 다음은 Xgboost는 모델을 학습시키기 위한 다양한 Parameter가 존재합니다. 해당하는 값의 문제를 해결해주기 위해 첫번째로는 정확도를 높이는 방식으로 튜닝을 진행해주고, 이후에 내용을 학습하였다

## Step1. 정확도 향상
- max_depth, min_child_weight, n_estimator 하이퍼파라미터 튜닝

```python
# 정확도 향상
from sklearn.model_selection import GridSearchCV

params = {
 'max_depth':range(3,20,3),
 'min_child_weight':range(1,6,2),
 'n_estimators':range(100,1100,100)
}
grid_xgb = GridSearchCV(estimator = xgboost.XGBClassifier(learning_rate=0.1, 
                                                         n_estimators=1000, 
                                                         max_depth=5, 
                                                         min_child_weight=1, 
                                                         gamma=0, 
                                                         subsample=0.8, 
                                                         colsample_bytree=0.8,
                                                         objective= 'binary:logistic', 
                                                         nthread=-1, 
                                                         scale_pos_weight=1, 
                                                         seed=2019),
                        param_grid = params, n_jobs=-1)
grid_xgb.fit(train_x,train_y)
print(grid_xgb.best_params_, grid_xgb.best_score_)
```

### Step2
- gamma 하이퍼파라미터 튜닝

```python
from sklearn.model_selection import GridSearchCV

params = {
    'gamma':[i/10.0 for i in range(0,5)]
}
grid_xgb = GridSearchCV(estimator = xgboost.XGBClassifier(learning_rate=0.1, 
                                                         n_estimators=100, 
                                                         max_depth=18, 
                                                         min_child_weight=1, 
                                                         gamma=0, 
                                                         subsample=0.8, 
                                                         colsample_bytree=0.8,
                                                         objective= 'binary:logistic', 
                                                         nthread=-1, 
                                                         scale_pos_weight=1, 
                                                         seed=2019),
                        param_grid = params, n_jobs=-1)
grid_xgb.fit(train_x,train_y)
print(grid_xgb.best_params_, grid_xgb.best_score_)
```

### Step3

```python
from sklearn.model_selection import GridSearchCV

params = {
    'colsample_bytree':[i/10.0 for i in range(6,10)],
    'subsample':[i/100.0 for i in range(40,100)],
    
}
grid_xgb = GridSearchCV(estimator = xgboost.XGBClassifier(learning_rate=0.1, 
                                                         n_estimators=100, 
                                                         max_depth=18, 
                                                         min_child_weight=1, 
                                                         gamma=0, 
                                                         subsample=0.8, 
                                                         colsample_bytree=0.8,
                                                         objective= 'binary:logistic', 
                                                         nthread=-1, 
                                                         scale_pos_weight=1, 
                                                         seed=2019),
                        param_grid = params, n_jobs=-1)
grid_xgb.fit(train_x,train_y)
print(grid_xgb.best_params_, grid_xgb.best_score_)
```

### 최종 XGboost Regressor에 체크

```python
from sklearn.metrics import roc_auc_score
import tqdm
best_seed = []
for i in tqdm.notebook.tqdm(range(2019)):
    xgb_model=xgboost.XGBClassifier(learning_rate=0.1,
                     n_estimators=100,
                     max_depth=18,
                     min_child_weight=1,
                     gamma=0.0,
                     subsample=0.6,
                     colsample_bytree=0.99,
                     objective= 'binary:logistic',
                     nthread=-1,
                     scale_pos_weight=1,
                     reg_alpha = 1,
                     seed=i)
    xgb_model.fit(train_x,train_y)
    best_seed.append((roc_auc_score(list(train_y), xgb_model.predict(train_x)),i))
```
